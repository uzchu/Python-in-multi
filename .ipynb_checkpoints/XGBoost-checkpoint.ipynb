### xgboost 적용

```python
import xgboost as xgb
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_features, y_label,test_size=0.2, random_state=0)
```

```python
type(X_train)
```

```python
dtrain = xgb.DMatrix(data=X_train, label=y_train)
dtest = xgb.DMatrix(data=X_test, label=y_test)

type(dtrain)
```

```python
#웬만하면 아래 params는 한 세트라고 생각하기
params = {'max_depth':3, # 4-5를 넘기지 않음, 트리의 최대 깊이값, 적절하게 조정하며 accuracy 테스트해야함, 얘를 줄이는 대신 반복을 늘림
         'eta':0.1,   # 진행방향의 간격값 (LearningRate값) 크면 간격이 넓어서 빠르게 구할 수 있음(대신 건너뛸수도 있음)
                      #기울기 앞에 붙는 상수값, 경사하강법식에서 n인지 처럼 생긴거, https://blog.naver.com/ollehw/221517423957
          'objective':'binary:logistic', #목적, 분류모형
          'eval_metric':'logloss',  #cost_function에 해당하는 부분, 예측치-실측치
          'early_stoppings':100 #이걸 수행하기 위해선 반드시 위에 eval_metric 옵션을 꼭 줘야함.
         }
num_rounds = 643 #부스팅의 반복횟수
wlist = [(dtrain,'train'),(dtest,'eval')] #dtest가 여기선 test(newdata)가 아니고 평가데이터, 사실 dvalidation이 더 정확함..
xgb_model = xgb.train(params = params, \
                     dtrain = dtrain, num_boost_round = num_rounds, evals=wlist) #evals 평가옵션

#이전까지 accuracy로 보여주던걸 손실율(logloss)로 보여줌
#손실율을 줄여가야 함
#eval-logloss : 평가지표에 대해 얼마만큼 잔차를 작게 만드느냐
```

```python
pred_probs = xgb_model.predict(dtest) #얘가 new데이터
print('predict() 수행결과값을 10개만 표시, 예측 확률 값으로 표시됨')
print(np.round(pred_probs[:10],3))
```

```python
#예측 확률이 0.5보다 크면 1, 그렇지 않으면 0으로 예측값 결정하여 List 객체인 preds
preds = [1 if x>0.5 else 0 for x in pred_probs]
print('예측값 10개만 표시 :', preds[:10])
```
